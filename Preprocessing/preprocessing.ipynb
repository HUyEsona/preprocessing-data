### Chuyển đổi kiểu dữ liệu

Như có thể thấy ở đây một vài dữ liệu số đang ở type là object vì vậy ta cần đưa chúng về kiểu dữ liệu số int64 hoặc float64

```python
columns_to_int = [
    'Diện Tích Lãnh Thổ (km²)',
    'Tổng Dân Số',
    'Mật Độ Dân Số (người/km²)',
    'Tuổi Thọ Trung Bình',
    'Số Bang/Tỉnh',
    'Số Di Sản Văn Hoá UNESCO'
]

columns_to_float = [
    'Tỷ Lệ Đô Thị Hoá (%)', 
    'Tỷ Lệ Biết Chữ (%)', 
    'Tỷ Lệ Sinh (‰)', 
    'Tỷ Lệ Tử (‰)',
    'GDP (triệu tỷ USD)', 
    'GDP Bình Quân Đầu Người (người/USD)', 
    'Tỷ Lệ Thất Nghiệp (%)',
    'Tổng Số Khách Đến Du Lịch (triệu người)',
    'Tổng Thu Nhập Du Lịch (tỷ USD)',
    'Chỉ Số HDI (Human Development Index)'
]
```

```python
# Xử lý cột chuyển đổi sang kiểu int: 
for col in columns_to_int:
    df[col] = df[col].astype(str).str.replace(',', '')   # xoá dấu phẩy
    df[col] = df[col].str.strip()                        # xoá khoảng trắng đầu/cuối
    df[col] = pd.to_numeric(df[col], errors='coerce')    # ép kiểu float64 
    df[col] = df[col].round().astype("Int64")
```

```python
df
```

```python
for col in columns_to_float:
    df[col] = df[col].astype(str).str.replace(',', '', regex=False)
    df[col] = df[col].str.replace('%', '', regex=False)
    df[col] = df[col].str.strip()
    df[col] = pd.to_numeric(df[col], errors='coerce')

    # Nếu giá trị lớn hơn 1 nhưng dưới 200 → chia 100 để thành phần trăm
    if df[col].max(skipna=True) > 1 and df[col].max(skipna=True) <= 200:
        df[col] = df[col] / 100

    # Làm tròn đến 2 chữ số sau dấu phẩy
    df[col] = df[col].round(2)
```

```python
df_null_info = df[['Tên Quốc Gia']].copy()
df_null_info['null_count'] = df.isna().sum(axis=1)

# Lọc những dòng có ít nhất 1 NaN
df_null_info = df_null_info[df_null_info['null_count'] > 0]

# Sắp xếp theo null_count giảm dần
df_null_info = df_null_info.sort_values(by='null_count', ascending=False)

# In ra
print(df_null_info)
```

Dữ liệu đang chứa khá nhiều giá trị NULL và đặc biệt là với cột `Tổng Số Lượng Khách Du Lịch Nước Ngoài` là : 44/195 

```python
# Xem các dòng có giá trị null
df_missing = df[df.isna().any(axis=1)]
df_missing
```

```python
# Xoá các hàng có hơn 50% giá trị null
df = df.loc[df.isnull().mean(axis=1) < 0.5]
```
